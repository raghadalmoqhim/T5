## 1. Exploratory Data Analysis (EDA)

### Perform summary statistics on the dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv("/content/drive/MyDrive/T5/WEEK_1/exam_day5/Task1/Task_(2)_Dataset.csv")
```

### Identify and analyze patterns in the data.

```python
print(df.isnull().sum())
```

```python
# Analyze the relationship between TrafficIndexLive and Jamscount
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TrafficIndexLive', y='JamsCount', data=df)
plt.title('Relationship between TrafficIndexLive and Jamscount')
plt.xlabel('Traffic Index')
plt.ylabel('Number of Jams')
plt.show()

# Explore the trend of TrafficIndexLive over time
df['Datetime'] = pd.to_datetime(df['Datetime'])
df.set_index('Datetime', inplace=True)
plt.figure(figsize=(12, 6))
df['TrafficIndexLive'].plot()
plt.title('Traffic Index Over Time')
plt.xlabel('Date')
plt.ylabel('Traffic Index')
plt.show()

# Compare TrafficIndexLive with TrafficIndexWeekAgo
plt.figure(figsize=(10, 6))
sns.boxplot(x='City', y='TrafficIndexLive', data=df)
plt.title('Traffic Index by City')
plt.xticks(rotation=45)
plt.show()
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
# Visualize the distribution of TrafficIndexLive
plt.figure(figsize=(10, 6))
sns.histplot(df['TrafficIndexLive'], kde=True)
plt.title('Distribution of TrafficIndexLive')
plt.xlabel('Traffic Index')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualize the distribution of Datetime
plt.figure(figsize=(12, 6))
df.index.value_counts().sort_index().plot(kind='line')
plt.title('Distribution of Datetime')
plt.xlabel('Date')
plt.ylabel('Frequency')
plt.show()
```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python
# relationships between variables (Traffic_Index vs. Weather_Condition)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TrafficIndexLive', y='JamsDelay', data=df)
plt.title('Relationship between TrafficIndexLive and JamsDelay')
plt.xlabel('Traffic Index')
plt.ylabel('Jams Delay (minutes)')
plt.show()
```
